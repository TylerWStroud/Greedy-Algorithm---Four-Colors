```python
#
# Question 2: Graph Coloring Greedy (Four-Color Theorem)
#

def greedy_graph_coloring(adjacency, num_colors=4):
    """
    Assign colors to nodes using a greedy algorithm.
    For each node, pick the smallest color (1..num_colors)
    not already used by an adjacent node.

    Parameters:
        adjacency (dict): {node: [neighbor, ...]} adjacency list.
        num_colors (int): Maximum number of colors allowed.

    Returns:
        color_map (dict): {node: color_number}
    """
    color_map = {}  # Stores color assignment for each node

    for node in adjacency:
        # Collect colors already used by neighbors
        neighbor_colors = {color_map[nb] for nb in adjacency[node]
                           if nb in color_map}

        # Assign the lowest available color
        for color in range(1, num_colors + 1):
            if color not in neighbor_colors:
                color_map[node] = color
                break
        else:
            raise ValueError(f"Could not color '{node}' with {num_colors} colors.")

    return color_map


def verify_coloring(adjacency, color_map):
    """Verify no two adjacent states share the same color."""
    for node, neighbors in adjacency.items():
        for nb in neighbors:
            if color_map[node] == color_map[nb]:
                return False, f"Conflict: {node} and {nb} both have color {color_map[node]}"
    return True, "Valid – no two adjacent states share a color."


# US Contiguous-State Adjacency List
us_adjacency = {
    "WA": ["OR", "ID"],
    "OR": ["WA", "ID", "NV", "CA"],
    "CA": ["OR", "NV", "AZ"],
    "ID": ["WA", "OR", "NV", "UT", "WY", "MT"],
    "NV": ["OR", "CA", "ID", "UT", "AZ"],
    "AZ": ["CA", "NV", "UT", "NM"],
    "UT": ["ID", "NV", "AZ", "CO", "WY"],
    "MT": ["ID", "WY", "ND", "SD"],
    "WY": ["MT", "ID", "UT", "CO", "NE", "SD"],
    "CO": ["WY", "UT", "NM", "OK", "KS", "NE"],
    "NM": ["AZ", "UT", "CO", "OK", "TX"],
    "ND": ["MT", "SD", "MN"],
    "SD": ["ND", "MT", "WY", "NE", "IA", "MN"],
    "NE": ["SD", "WY", "CO", "KS", "MO", "IA"],
    "KS": ["NE", "CO", "OK", "MO"],
    "OK": ["KS", "CO", "NM", "TX", "AR", "MO"],
    "TX": ["NM", "OK", "AR", "LA"],
    "MN": ["ND", "SD", "IA", "WI"],
    "IA": ["MN", "SD", "NE", "MO", "IL", "WI"],
    "MO": ["IA", "NE", "KS", "OK", "AR", "TN", "KY", "IL"],
    "AR": ["MO", "OK", "TX", "LA", "MS", "TN"],
    "LA": ["TX", "AR", "MS"],
    "WI": ["MN", "IA", "IL", "MI"],
    "IL": ["WI", "IA", "MO", "KY", "IN"],
    "MI": ["WI", "IN", "OH"],
    "IN": ["IL", "MI", "OH", "KY"],
    "OH": ["MI", "IN", "KY", "WV", "PA"],
    "KY": ["IL", "MO", "TN", "VA", "WV", "OH", "IN"],
    "TN": ["MO", "AR", "MS", "AL", "GA", "NC", "VA", "KY"],
    "MS": ["AR", "LA", "TN", "AL"],
    "AL": ["MS", "TN", "GA", "FL"],
    "GA": ["AL", "TN", "NC", "SC", "FL"],
    "FL": ["AL", "GA"],
    "SC": ["GA", "NC"],
    "NC": ["TN", "VA", "SC", "GA"],
    "VA": ["KY", "TN", "NC", "WV", "MD", "DC"],
    "WV": ["KY", "VA", "MD", "PA", "OH"],
    "PA": ["OH", "WV", "MD", "DE", "NJ", "NY"],
    "MD": ["VA", "WV", "PA", "DE", "DC"],
    "DC": ["VA", "MD"],
    "DE": ["MD", "PA", "NJ"],
    "NJ": ["PA", "DE", "NY"],
    "NY": ["PA", "NJ", "CT", "MA", "VT"],
    "CT": ["NY", "RI", "MA"],
    "RI": ["CT", "MA"],
    "MA": ["NY", "CT", "RI", "NH", "VT"],
    "VT": ["NY", "MA", "NH"],
    "NH": ["VT", "MA", "ME"],
    "ME": ["NH"]
}

COLOR_NAMES = {1: "Red", 2: "Green", 3: "Blue", 4: "Yellow"}

# RUN
color_assignment = greedy_graph_coloring(us_adjacency, num_colors=4)

# Print lookup table
print("=" * 50)
print(" Graph Coloring – State Color Lookup Table")
print("=" * 50)
print(f"{'State':<8} {'Color #':<10} {'Color Name'}")
print("-" * 30)
for state, color_num in sorted(color_assignment.items()):
    print(f"{state:<8} {color_num:<10} {COLOR_NAMES[color_num]}")

# Verify correctness
valid, message = verify_coloring(us_adjacency, color_assignment)
print()
print("Verification:", message)
colors_used = len(set(color_assignment.values()))
print(f"Total colors used: {colors_used} / 4")
```
